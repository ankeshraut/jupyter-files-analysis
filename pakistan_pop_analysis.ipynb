```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# data import
df=pd.read_csv('Titanic-Dataset.csv')
df
```

```python
# check the data
df.head()
```

```python
# find out the percentage of missing values
df.isnull().sum()/len(df)*100
```

```python
#1- drop the age column from dataset having more than 70% of missing values from dataset
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.columns
```

```python
df['Age'].mean()
```

```python
df['Age'].median()
```

```python
df['Age'].mode()[0]
```

```python
# 2- filling missing values of age column by the mean of age 
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

# feature engineering

```python
# feature engineering means that add the new column in dataset
```

```python
# binning of age column into 7 categories
bins = [0,1,5,12,18,30,50,80]
labels=['Infants','Toddlers','Kids','Teens','Youngs','Middle Aged','Old']
# CONVERT BINS INTO COLUMNS
df['binned_age']=pd.cut(df['Age'],bins=bins,labels=labels)
```
